```python
# 1. Text Analysis Tool

import re
from collections import Counter

# Function to count the words in the text.
def word_count(text):
    words = text.split()
    return len(words)

# Function to count the sentences in the text.
def sentence_count(text):
    sentences = re.split(r'[.!?]', text)
    return len(sentences)

# Function to count the paragraphs in the text (separated by two line breaks).
def paragraph_count(text):
    paragraphs = text.split('\n\n')
    return len(paragraphs)

# Function to calculate the average word length in the text.
def average_word_length(text):
    words = text.split()
    total_length = sum(len(word) for word in words)
    return total_length / len(words)

# Function to find the most common words in the text.
def most_common_words(text, n=10):
    words = re.findall(r'\b\w+\b', text.lower())
    word_counts = Counter(words)
    most_common_words = word_counts.most_common(n)
    return most_common_words

file_path = 'Text.txt'

try:
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
except FileNotFoundError:
    print(f"File '{file_path}' not found.")
else:
    # Call the functions and display the results.
    word_count = word_count(text)
    sentence_count = sentence_count(text)
    paragraph_count = paragraph_count(text)
    average_word_length = average_word_length(text)
    most_common_words = most_common_words(text)

    print(f"Words: {word_count}")
    print(f"Sentences: {sentence_count}")
    print(f"Paragraphs: {paragraph_count}")
    print(f"Average word length: {average_word_length:.2f} characters")

    print("\nThe Most common words:")
    for word, frequency in most_common_words:
        print(f"{word}: {frequency}")
```

```python
# 2. Hangman Game

import random

# Word list of the game
word_list = ["pizza", "hamburger", "tacos", "soda", "chips"]

# Function to initialize the game state
def choose_word():
    return random.choice(word_list)

# Function to initialize the game state
def initialize_game():
    word = choose_word()
    guessed_letters = []
    attempts = 6
    return word, guessed_letters, attempts

# Function to show the current state of the word with underscores for hidden letters
def display_word(word, guessed_letters):
    display = ""
    for letter in word:
        if letter in guessed_letters:
            display += letter
        else:
            display += "_"
    return display

# Function to play Hangman
def play_hangman():
    word, guessed_letters, attempts = initialize_game()
    print("Hangman Game!")

    while attempts > 0:
        print("\nAttempts left:", attempts)
        current_display = display_word(word, guessed_letters)
        print("Current word:", current_display)

        if current_display == word:
            print("Congratulations! You guessed the word:", word)
            break

        guess = input("Guess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a valid single letter.")
            continue

        if guess in guessed_letters:
            print("You've already guessed that letter.")
            continue

        guessed_letters.append(guess)

        if guess in word:
            print("Good guess!")
        else:
            print("Wrong guess.")
            attempts -= 1

    if attempts == 0:
        print("Game over. The word was:", word)

# Start the game
play_hangman()
```

```python
# Number Guessing Game

import random

# Generate a random number between 1 and 100
number = random.randint(1, 100)

# Number of guesses
attempts = 0

print("Guess the Number GAME!")

# Start a loop that continues until the user guesses the correct number
while True:
    try:
        # Ask for the number
        guess = int(input("Guess the number between 1 and 100: "))

        # Increment the number of attempts
        attempts += 1

        # Check if the user's guess is correct
        if guess == number:
            print(f"Congratulations! You guessed the number {number} in {attempts} attempts!")
            break  # Exit the loop when the correct number is guessed
        elif guess < number:
            print("The number is higher.")
        else:
            print("The number is lower.")
    except ValueError:
        print("Invalid input. Enter a number between 1 and 100.")
```
